```python
from scipy.optimize import curve_fit; import numpy as np; reload(material_analytics)

data22 = strainmodel('ref/HSRS/22').get_experimental_data()
yield1 = material_analytics.yield_stress(data22,cutoff=0.25)

elastic = data22[data22[:,0]<yield1[0,0]]
plastic = data22[data22[:,0]>=yield1[0,0]]

#plot.plotmult2D(elastic,plastic,marker1='bo',marker2='ro')

def quad(x,a,b):
    return a*x**2+b

x_dom = np.linspace(0.,yield1[0,0],1000)

optimal_params, cov_matrix = curve_fit(quad,elastic[0],elastic[1])
a, b = optimal_params

def bestquad(x):
    return a*x**2+b

pred_data = material_analytics.samplepoints(bestquad,[0,yield1[0,0]],1000)
#print pred_data
plot.plotmult2D(pred_data,elastic)
```
